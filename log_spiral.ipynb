```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
t_min = -20
t_max = -3
steps = 1000
xo = 1.17
yo = 0.28
k = 0.31
a = 2.93
```

```python
lin_a = np.linspace(t_min,t_max,1000)
print(lin_a)
```

```python
x = np.array(xo+(np.exp(k*lin_a))*(np.cos(lin_a)))
y = np.array(yo+ a*(np.exp(k*lin_a))*(np.sin(lin_a)))
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.title('Log Spiral')
plt.show()
```
