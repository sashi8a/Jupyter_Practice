```python
df_can.set_index('Country',inplace = True) #setting the indicies as the country names instead of numbers
```

```python
years = list(range(1980,2014))   #creates a list from years 1980 to 2013, (inclusive, exclusive)
```

```python
print(years)
```

```python
Albania = df_can.loc['Albania',years]
Algeria = df_can.loc['Algeria',years]
```

```python
Albania.plot(kind = 'line', color = 'green')
Algeria.plot(kind = 'line', color = 'red')

plt.show()
```

```python
difference = (Algeria-Albania)
plt.plot(years,difference, color = 'blue')
plt.show()
```
